```python
import matplotlib.pyplot as plt
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
import csv
import pandas as pd
```

```python
import numpy as np
```

```python
sev=1376
non_sev=4324
sev_p=[0.884,0.711,0.603,0.442,0.376,0.357,0.26,0.26,0.161,0.113,0.078,
       0.059,0.057,0.028]
non_sev_p=[0.814,0.657,0.442,0.057,0.28,0.128,0.131,0.109,0.121,0.135,0.097,
           0.057,0.058,0.051]
sev_n=[round(sev*i) for i in sev_p]
non_sev_n=[round(non_sev*i) for i in non_sev_p]
```

```python
sym=['Fever','Cough','Fatigue','Dyspnea','Sputum production','Shortness of breath'
     ,'Myalgia','Chill','Dizziness','Headache','Sore throat','Nausea or vomiting',
    'Diarhoea','Nasal congestion']
```

```python
G=nx.Graph()
for i in range(1, len(sym)):
    G.add_edge(sym[i-1], sym[i], weight=sev_n[i-1]-sev_n[i])
G.add_edge('Fever','Nasal congestion',weight=sev_n[-1]-sev_n[0])
```

```python
pos = graphviz_layout(G)
plt.axis('off')
plt.figure(figsize=(8,8))
plt.margins(x=0.2,y=0.2)
nx.draw_networkx_nodes(G, pos, node_color='g')  # draws nodes
nx.draw_networkx_edges(G, pos, edge_color='r')  # draws edges
nx.draw_networkx_labels(G, pos)  # node lables
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos,edge_labels=labels)
plt.show()
```
